Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 

```python
medv = np.array(data.MEDV).reshape(-1,1)
lstat = np.array(data.LSTAT).reshape(-1,1)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    medv,
    lstat,
    test_size = 0.2
)
```

```python
# Обучение
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
predicted_y = model.predict(X_test)

print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
# Построим график
plt.xlabel('freedom')
plt.ylabel('positive affect')
plt.title('Линейная регрессия от одной переменной')
x = np.linspace(-100, 200, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(-0,60)
plt.ylim(-5,50)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
import pylab
import numpy

iterat = 100
val = numpy.random.randint(1,7,size=iterat)

i = 0
while i < 1000:
    val=val+numpy.random.randint(1,7,size=iterat)
    i = i + 1
pylab.hist(val/1000, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5], edgecolor='blue')

pylab.xlabel('Значение')
pylab.ylabel('Количество')
pylab.title('Игральный кубик: ' + str(iterat) + ' подбрасываний')
pylab.show()
```

```python
# Потому что наиболее часто встречающаяся величина - 4 (так же было в 10 подбрасываний). 
# Распределение стремится к равномерному, стреднее значение остается таким же. 
# Можно применить центральную предельную теорему, так как испытаний много (1000)
```
